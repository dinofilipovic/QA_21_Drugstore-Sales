```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('gdrive/My Drive/Algebra_MBA/Quantitative_Analysis/Practice/21_Drugstore Sales.csv')
```

```python
correlations = df.corr()
```

```python
print(correlations)
```

```python
# Fit a linear regression model
model = LinearRegression()
model.fit(df[['Sales']], df['Promote'])

# Get the coefficients
a = model.coef_[0]
b = model.intercept_
```

```python
predicted_runs = a * df['Sales'] + b
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales'], df['Promote'], color='blue', label='Sales')

plt.xlabel('Sales')
plt.ylabel('Promote')
plt.title('Scatterplots of Sales and Promotions')
plt.legend()
plt.grid(True)

# Display Regression Line
plt.plot(df['Sales'], predicted_runs, color='red', label='Regression Line')

# Display Plot
plt.show()
```
